```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

22.1 Ordinary differential equations are a class of variables where any level of derivative is related to either another rate of change or the value of the variables.  These are important throughout most engineering and physics disciplines.  While some can be solved analytically, most are only solvable with numerical methods.  

An ordinary differential equations is different from a partial differential equation based on the number of variables effecting the rates,  An ordinary equation only has 1 variable.



```python
g=9.8
l=2
def acceleration(theta):
  return -1*g/l*np.sin(theta)

print('The acceleration of the pendulum is directly influenced by the position in the swing')
theta = np.linspace(-3.14,3.14,1000)
plt.plot(theta, acceleration(theta))
plt.ylabel('d$^2$Theta/dt$^2$')
plt.xlabel('theta')
plt.show()
```

22.2 Reduction of order is based on the fact that if the ODE can be expressed as a linear combination of the derivatives of the state, the problem become infinitely easier to solve.  Before blindly applying numerical methods, the problem should be analysed to determine if we are working with a solvable system.

22.3 A simple way to solve an ODE is to use the Euler method, which does a numerical integration of the values, based on the current value.  

```python
dt=.1
theta0=2
maxPoints =int(10/dt)

F=np.zeros(maxPoints)
theta=theta0
dtheta = 0
t=np.zeros(maxPoints)
for i in range(maxPoints):
  atheta = acceleration(theta)*dt 
  dtheta += atheta*dt
  theta += dtheta*dt
  F[i]=theta
  t[i]=i*dt

plt.plot(t, F, label='Numerical Solution')  
plt.plot(t, 2*np.cos(t*(g/l)**.5), label='Analytic Solution')  #*.235
plt.xlabel('Time')
plt.ylabel('Theta')
plt.legend()
plt.show()
```

It is easy to see that the error accumulates quickly with this system. A smaller time step can help, but advanced techniques are often needed to prevent brek down of the system.
